```python
# FIND ELF WITH MOST AMOUNT OF CALORIES

f = open("day1.txt")
data = f.readlines()
f.close()

running_tally = 0
list_of_answers = []

for i,el in enumerate(data):
    if el == "\n":
        print("Skip detected")
        print(f"Running Tally at: {running_tally}")
        list_of_answers.append(running_tally)
        print(f"Pushing Running Tally to List of Answers: {list_of_answers}")
        running_tally = 0
        print(f"Flushing running tally: {running_tally}")
        print("Starting Again!")
    

    else:
        #print(el,type(el),len(el))
        el = el.replace("\n","")
        #print(el, type(el),len(el))
        running_tally += int(el)
        print(f"RUNNING TALLY AT {running_tally} WITH ADDITION OF {el}")

sorted(list_of_answers)[-1]
```

```python
# FIND TOP THREE ELFS WITH MOST AMOUT OF CALORIES AND SUM

list_to_sum = sorted(list_of_answers)
#sum(list_to_sum[-2:]) + 68218 # 204837, right answer!
print(list_to_sum[-1] + list_to_sum[-2] + list_to_sum[-3])
# 203194 is too low!
```
